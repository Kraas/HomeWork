```python
# Загрузить файл wines_data.csv
# Рассчитать и визуализировать корреляции для переменных данного датасета
# Визуализировать взаимосвязи между 3-4 переменными
# Рассчитать t-test статистики для 3-4 переменных
```

```python
import pandas as pd
from scipy import stats
import seaborn as sns
```

```python
data = pd.read_csv('wine_data.csv')
```

```python
data.head()
```

```python
corr = data.corr()
corr
```

```python
sns.heatmap(corr, annot=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
sns.pairplot(data, vars=['Flavanoids', 'Total phenols'],
                 kind='scatter')
```

```python
sns.pairplot(data, vars=['Alcohol', 'Proline'],
                 kind='scatter')
```

```python
sns.pairplot(data, vars=['Class', 'Flavanoids'],
                 kind='scatter')
```

```python
sns.pairplot(data, vars=['Ash', 'Alcalinity of ash'],
                 kind='scatter')
```

```python
stats.ttest_ind(data['Alcohol'], data['Proline'])
```

```python
stats.ttest_ind(data['Class'], data['Flavanoids'])
```

```python
stats.ttest_rel(data['Ash'], data['Alcalinity of ash'])
```

```python
stats.ttest_1samp(data['Flavanoids'] - data['Total phenols'], 0)
```
